2) Obtenga la función transferencia, pero normalizada.
Para la normalizacion diremos que R1=R2 y que 1/R3C = K, pasando de (I) a (II).

(I)
\begin{equation}
T(S)= \frac{S-\frac{R2}{R1 R3 C}}{S+\frac{1}{R3 C}}
\end{equation}

(II)
\begin{equation}
T(S)= \frac{S-K}{S+K}
\end{equation}

Luego remplazo que S=KP, dando
\begin{equation}
T(P)= \frac{KP-K}{KP+K}
\end{equation}

Quedando
\begin{equation}
T(P)= \frac{K(P-1)}{K(P+1)}
\end{equation}

Finalmente la transformada normalizada :
\begin{equation}
T(P)= \frac{P-1}{P+1}
\end{equation}
