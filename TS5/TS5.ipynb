```python
import scipy
import scipy.signal as signal
from scipy import signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq


Qz=10000
Qp=1/(2**(1/2))
Wp=0.05
Wz=0.1
K=1

num=[1*K,K*(Wz/Qz),K*Wz**2]
den=[1,(Wp/Qp),Wp**2]

mi_sos_A = sig.TransferFunction(num,den)
analyze_sys(mi_sos_A, 'Caso A')



Wp=1
Wz=1.99526
num=[1,(Wz/Qz),Wz**2]
den=[1,(Wp/Qp),Wp**2]

mi_sos_B = sig.TransferFunction(num,den)
analyze_sys(mi_sos_B, 'Caso B')

Qz=-Qp
Wp=1
Wz=1
num=[1,(Wz/Qz),Wz**2]
den=[1,(Wp/Qp),Wp**2]

mi_sos_B = sig.TransferFunction(num,den)
analyze_sys(mi_sos_B, 'Caso C')


Qz=10000
Qp=1/(2**(1/2))
Wp=0.5
Wz=1

num=[1,(Wz/Qz),(Wz**2)]
den=[1,(Wp/Qp),Wp**2]

mi_sos_B = sig.TransferFunction(num,den)
analyze_sys(mi_sos_B, 'Caso D')
```
